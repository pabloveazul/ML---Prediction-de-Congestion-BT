```python
def getNonNumericalColumns(dataset):
    liste = []
    for index in range(len(dataset.dtypes)):
        if not((dataset.dtypes[index] == 'float64') or (dataset.dtypes[index] == 'int64')):
            liste.append(index)
    return(liste)
```

```python
def prepareData(dataset):
    non_numerical = getNonNumericalColumns(dataset)
    values = dataset.values
    #Encode non numerical Data (TEC)
    encoder = LabelEncoder()
    encoder.fit(values[:,non_numerical[0]])
    for index in non_numerical:
        values[:,index] = encoder.transform(values[:,index])
    values = values.astype('float32')
    #Scale Data
    scaler = MinMaxScaler()
    values = scaler.fit_transform(values)
    return(values)
```

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = pd.DataFrame(data)
	cols, names = list(), list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
		names += [(name+'(t-%d)' % (i)) for name in data.columns]
	# forecast sequence (t, t+1, ... t+n)
	for i in range(0, n_out):
		cols.append(df.shift(-i))
		if i == 0:
			names += [name+'(t)' for name in data.columns]
		else:
			names += [(name+'(t+%d)' % (i)) for name in data.columns]
	# put it all together
	agg = pd.concat(cols, axis=1)
	agg.columns = names
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg
```

```python
def generateGeneralSupervisedData(n_predictive_weeks,n_future_weeks):
    groupe= '_all_'
    data = 'all.xlsx'
    local_path = 'C:/Users/euripab/Desktop/ML Extraction/'
    #Creating General Supervised Data
    liste_MeanMaxMin_PerDayData = os.listdir(local_path + "/Statistics/results1_MeanMaxMin_PerDay/")
    liste_ClusteringTimeSeriesData = os.listdir(local_path + "/KSTest/results5_ClusteringTimeSeries/")
    liste_cells = pd.read_excel(r'C:\Users\euripab\Desktop\ML Extraction\KSTEST\results3_PvaluesDays\KSTest_2018-08-07_all.xlsx').index
    allCellSupervisedData = pd.DataFrame()
    predictive_time_units = n_predictive_weeks*7
    to_predict_time_units = (n_future_weeks-1)*7+1
    for cell in liste_cells:
        for current_data in liste_MeanMaxMin_PerDayData:
            if (groupe in current_data) and (cell in current_data):
                timeSeries_MeanMaxMin_PerDay = pd.read_excel(local_path + "/Statistics/results1_MeanMaxMin_PerDay/" + current_data)
                timeSeries_MeanMaxMin_PerDay = timeSeries_MeanMaxMin_PerDay.reset_index(drop=True)
                break
        for current_data in liste_ClusteringTimeSeriesData:
            if (groupe in current_data) and (cell in current_data):
                timeSeries_ClusteringTimeSeries = pd.read_excel(local_path + "/KSTest/results5_ClusteringTimeSeries/" + current_data)
                timeSeries_ClusteringTimeSeries = timeSeries_ClusteringTimeSeries.reset_index(drop=True)
                break
        timeSeries = timeSeries_ClusteringTimeSeries.join(timeSeries_MeanMaxMin_PerDay, how='outer')
        cellSupervisedData = series_to_supervised(timeSeries, n_in=predictive_time_units, n_out=to_predict_time_units, dropnan=True)
        allCellSupervisedData = allCellSupervisedData.append(cellSupervisedData, ignore_index=True)
    fic_result= local_path + "/General_SupervisedData/results1_MeanMaxMin_Clustering/" +'supervised' +groupe+str(predictive_time_units)+'_'+str(to_predict_time_units)+'.xlsx'
    writer_results = pd.ExcelWriter(fic_result)
    allCellSupervisedData.to_excel(writer_results)
    writer_results.save()
    print('Supervised '+str(predictive_time_units)+' - '+str(to_predict_time_units)+' done')
    return()
```

```python
def generalSupervisedDataGenerator(liste_predictive_weeks,liste_future_weeks):
    groupe= '_all_'
    local_path = 'C:/Users/euripab/Desktop/ML Extraction/'
    liste_ClusteringTimeSeriesData = os.listdir(local_path+ "/General_SupervisedData/results1_MeanMaxMin_Clustering/")
    for n_predictive_weeks in liste_predictive_weeks:
        predictive_time_units = n_predictive_weeks*7
        
        for n_future_weeks in liste_future_weeks:
            to_predict_time_units = (n_future_weeks-1)*7+1
            if 'supervised' +groupe+str(predictive_time_units)+'_'+str(to_predict_time_units)+'.xlsx' not in liste_ClusteringTimeSeriesData:
                generateGeneralSupervisedData(n_predictive_weeks,n_future_weeks)
            else:
                print('already done')
    return()
```

```python
def neuronalNetworkPredictor(n_predictive_weeks,n_future_weeks,kpi,data_dic):
    #read Data set
    dataset = data_dic[(n_predictive_weeks,n_future_weeks)]
    if dataset.empty:
        return((0,0,0))
    #Mix Data set
    dataset = dataset.sample(frac=1, replace=False)
    n_features = 29
    values = prepareData(dataset)
    dictionary = dict(zip(dataset.columns, range(0,len(dataset.columns))))
    dictionary_inv = dict(zip(range(0,len(dataset.columns,)),dataset.columns))
    #Use Information of one day from different weeks
    feat_list = [(i-1)*7*n_features+j  for i in range(1,n_predictive_weeks +1) for j in range(0,n_features)]
    #Use all Information available till the date
    #feat_list = range(0,n_features*n_predictive_weeks*7)
    print('Predictors :')
    for index in feat_list:
        print(dictionary_inv[index])
    dim = len(feat_list)
    
    #split Training and Validation Set
    n_train_samples = int(np.floor(len(values)/4))
    train = values[:n_train_samples, :]
    test = values[n_train_samples:, :]
    
    # split into input and outputs
    if (n_future_weeks-1) == 0:
        time_str = '(t)'
    else:
        time_str = '(t+'+str((n_future_weeks-1)*7)+')'
    print('prediction: '+kpi+time_str)
    
    train_X, train_y = train[:, feat_list], train[:,dictionary[kpi+time_str]]
    test_X, test_y = test[:, feat_list], test[:,dictionary[kpi+time_str]]
    
    # design network
    model = Sequential()
    model.add(Dense(dim*10, activation='relu', input_dim =dim))
    model.add(Dense(dim*2, activation='sigmoid'))
    model.add(Dense(1,activation = 'relu'))
    model.compile(loss='mse', optimizer='adam')
    # fit network
    model.fit(train_X, train_y, epochs=50, batch_size=72, validation_data=(test_X, test_y),verbose = 0, shuffle=False)
    #return loss in train and test
    return((model.evaluate(train_X, train_y),model.evaluate(test_X, test_y),model))

    
    
    
```

```python
def loadSupervisedData(liste_predictive_weeks,liste_future_weeks):
    out = {}
    print("Generating Supervised Data...")
    generalSupervisedDataGenerator(liste_predictive_weeks,liste_future_weeks)
    print("Loading Data...")
    for n_predictive_weeks in liste_predictive_weeks:
        for n_future_weeks in liste_future_weeks:
            out[(n_predictive_weeks,n_future_weeks)] = read_excel(r'C:\Users\euripab\Desktop\ML Extraction\General_SupervisedData\results1_MeanMaxMin_Clustering\supervised_all_'+str(n_predictive_weeks*7)+'_'+str((n_future_weeks-1)*7+1)+'.xlsx')
    print('Done!')
    return(out)
```

```python
data = loadSupervisedData([1,2,3,4],[1,2,3,4])
```

```python
def testNeuronalNetworkOnKpi(liste_predictive_weeks,liste_future_weeks,kpi,data_dic):
    out = DataFrame()
    for n_predictive_weeks in liste_predictive_weeks:
            for n_future_weeks in liste_future_weeks:
                score_train = 100
                score_test = 100
                for i in range(0,5):
                    score = neuronalNetworkPredictor(n_predictive_weeks,n_future_weeks,kpi,data_dic)
                    if score[0] < score_train:
                        score_train = score[0]
                    if score[1] < score_test:
                        score_test = score[1]     
                out.loc[str(n_predictive_weeks),str(n_future_weeks)] = '('+str(score_train)+','+str(score_test)+')'
    return(out)
```

```python
res = testNeuronalNetworkOnKpi([1,2,3,4],[1,2,3,4],'CQImoyen max',data)
```
