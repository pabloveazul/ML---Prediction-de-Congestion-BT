```python
def autocorrelation_array(data, maxlag):
    auto = []
    for lag in range(0,maxlag+1):
            ac = data.autocorr(lag)
            auto.append(ac)
    return(auto)        
```

```python
def extractCellKpiFeatures(celluleName):
    kpiFeatures = pd.DataFrame()
    kpiFeatures['CellName'] = [celluleName]
    for kpi in column_kpiselection:
        data_ck = alldata_kpis.loc[alldata_kpis['CellName']==celluleName][['DAY_',kpi]]
        data_ck['DAY_'] = pd.to_datetime(data_ck['DAY_'],infer_datetime_format=True)
        data_ck.sort_values(by='DAY_')
        f, Pxx = signal.welch(data_ck[kpi], 1/3600, nperseg=552, scaling='spectrum')
        welch = pd.DataFrame()
        welch['Freq'] = f
        welch['Power'] = Pxx
        welch = welch.sort_values(by=['Power'], ascending = False)
        A = autocorrelation_array(data_ck[kpi],24*7)
        ac = pd.DataFrame()
        ac['Lag'] = range(0,len(A))
        ac['AC'] = [abs(a) for a in A]
        ac = ac.sort_values(by=['AC'], ascending = False)
        for i in range(1,6):
            kpiFeatures[kpi+' Lag'+str(i)] = ac.iloc[i,0]
            kpiFeatures[kpi+' AC'+str(i)] = ac.iloc[i,1]
        kpiFeatures[kpi+' Mean'] = [data_ck[kpi].mean()]
        kpiFeatures[kpi+' Std'] = [data_ck[kpi].std()]
        kpiFeatures[kpi+' Max'] = [data_ck[kpi].max()]
        for i in range(0,6):
            kpiFeatures[kpi+' Freq'+str(i)] = welch.iloc[i,0]
            kpiFeatures[kpi+' Power'+str(i)] = welch.iloc[i,1]
    return(kpiFeatures)
```

```python
import multiprocessing

def extractCellKpiFeatures(queue,celluleName):
    kpiFeatures = pd.DataFrame()
    kpiFeatures['CellName'] = [celluleName]
    for kpi in column_kpiselection:
        data_ck = alldata_kpis.loc[alldata_kpis['CellName']==celluleName][['DAY_',kpi]]
        data_ck['DAY_'] = pd.to_datetime(data_ck['DAY_'],infer_datetime_format=True)
        data_ck.sort_values(by='DAY_')
        f, Pxx = signal.welch(data_ck[kpi], 1/3600, nperseg=552, scaling='spectrum')
        welch = pd.DataFrame()
        welch['Freq'] = f
        welch['Power'] = Pxx
        welch = welch.sort_values(by=['Power'], ascending = False)
        A = autocorrelation_array(data_ck[kpi],24*7)
        ac = pd.DataFrame()
        ac['Lag'] = range(0,len(A))
        ac['AC'] = [abs(a) for a in A]
        ac = ac.sort_values(by=['AC'], ascending = False)
        for i in range(0,6):
            kpiFeatures[kpi+' Lag'+str(i)] = ac.iloc[i,0]
            kpiFeatures[kpi+' AC'+str(i)] = ac.iloc[i,1]
        kpiFeatures[kpi+' Mean'] = [data_ck[kpi].mean()]
        kpiFeatures[kpi+' Std'] = [data_ck[kpi].std()]
        kpiFeatures[kpi+' Max'] = [data_ck[kpi].max()]
        for i in range(0,6):
            kpiFeatures[kpi+' Freq'+str(i)] = welch.iloc[i,0]
            kpiFeatures[kpi+' Power'+str(i)] = welch.iloc[i,1]
    queue.put(kpiFeatures)
    exit(kpiFeatures)

t = time.time()

if __name__ == '__main__':
    jobs = []
    result = multiprocessing.Queue()
    for cell in  np.random.choice(alldata_kpis['CellName'].unique(),100):
        p = multiprocessing.Process(target=extractCellKpiFeatures, args=(result,cell,))
        jobs.append(p)
        p.start()
    for proc in jobs:
        proc.join()
elapsed = time.time() - t     
print(elapsed)  
```
